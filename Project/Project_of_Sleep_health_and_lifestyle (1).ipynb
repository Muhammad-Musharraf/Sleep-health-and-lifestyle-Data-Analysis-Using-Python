## **Overview**
# Sleep Health and Lifestyle Data Analysis

This project explores the relationship between sleep health, lifestyle habits, physical activity, stress levels, BMI, occupation, and various health indicators using the Sleep Health and Lifestyle Dataset.

The analysis begins with data cleaning and preprocessing, including handling missing values and checking for duplicates. After preparing the dataset, several real-world health and lifestyle questions are investigated through data filtering, aggregation, and visualization.

Key areas explored in this project include:

* Sleep duration and quality across different occupations.
* The impact of stress levels on sleep quality and sleep duration.
* Relationships between BMI categories, physical activity, and overall health.
* Analysis of sleep disorders such as insomnia and sleep apnea.
* Demographic insights based on age and gender.
* Health risk identification through blood pressure, heart rate, and stress indicators.
* Occupation-based comparisons of sleep patterns, activity levels, and stress levels.

Various visualizations were created using Matplotlib and Seaborn, including bar charts, line plots, histograms, pie charts, and pair plots to uncover meaningful trends and patterns within the data.


#**Data Filtering**

# **1.Health & Sleep Analysis**




### **Q1. Which people sleep less than 6 hours per day and have high stress levels (>7)?**



 * Real-life impact: Identifies individuals at risk of burnout and mental health issues.

```python
m1=df['Sleep Duration']<6
m2=df['Stress Level']>7
df[m1 & m2][['Person ID','Gender','Stress Level','Sleep Duration']]
```

###**Q2. How many overweight or obese people suffer from sleep disorders?**



* Real-life impact: Helps understand the relationship between obesity and sleep problems.

```python
m1=df['BMI Category']=='Overweight'
m2=df['BMI Category']=='Obese'
m3=df['Sleep Disorder']!='None'
df[m1 | m2][m3]['BMI Category'].value_counts()
```

### **Q3.Which occupations have the highest number of people with sleep disorders?**



* Real-life impact: Helps employers improve workplace wellness programs.

```python
x1=df['Sleep Disorder']!='None'
df[x1]['Occupation'].value_counts()
```

###**Q4. Among people aged 40+, who sleeps less than 7 hours?**



* Real-life impact: Aging populations require adequate sleep for disease prevention.


```python
a1=df['Age']>=40
a2=df['Sleep Duration']<7
df[a1 & a2][['Person ID','Gender','Age','Sleep Duration']]
```

# **2. Lifestyle & Fitness Analysis**











### **Q5.Do people who walk more than 10,000 steps have better sleep quality?**



* Real-life impact: Encourages physical activity for better sleep health.

```python
a1=df['Physical Activity Level']>10000
a2=df['Quality of Sleep']>5
df[a1 & a2][['Person ID','Gender','Physical Activity Level','Quality of Sleep']]


if [a1 & a2] == None:
  print('People who walk more than 10,000 steps have better sleep quality')

else:
  print('People who walk more than 10,000 steps have not better sleep quality')
```

###**Q6. Which individuals have low physical activity (<30) and high BMI?**



* Real-life impact: Identifies people at risk of lifestyle-related diseases.

```python
a1=df['Physical Activity Level'] <=30
a2=df['BMI Category']=='Overweight'
a3=df['BMI Category']=='Obese'
df[a1 & (a2 | a3)][['Person ID','Gender','Physical Activity Level','BMI Category']]



```



### **Q7. What is the average sleep quality of people who exercise more than 60 minutes daily?**



* Real-life impact: Measures the effectiveness of physical activity on sleep.

```python
z1=df['Physical Activity Level']>60
df[z1]['Quality of Sleep'].mean()
```

# **3. Medical Risk Analysis**











### **Q8.Which people have high blood pressure and sleep apnea?**



* Real-life impact: Early identification of cardiovascular risk factors.

```python
df['Blood Pressure'] = df['Blood Pressure'].str.replace('/', ' ')
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split(' ', expand=True)
df['Systolic BP'].unique()
```

```python
m1=df['Systolic BP'].astype(int)>120
m2=df['Diastolic BP'].astype(int)>80
m3=df['Sleep Disorder']=='Sleep Apnea'
df[m1 & m2 & m3][['Person ID','Gender','Systolic BP','Diastolic BP','Sleep Disorder']]
```


#**Q9.How many individuals have a heart rate above 80 and stress levels above 8?**



   * Real-life impact: Detects potential stress-related health concerns.

```python
a1=df['Heart Rate']>80
a2=df['Stress Level'] >=8
df[a1 & a2][['Person ID','Gender','Heart Rate','Stress Level']]
```


# **10. Which BMI category has the highest average stress level?**



    * Real-life impact: Helps understand links between weight and mental health.

```python
x1=df.groupby('BMI Category')['Stress Level'].mean()
x1.idxmax()
```

# **4. Demographic Analysis**










# **11. Do males or females report better sleep quality?**



    * Real-life impact: Gender-based health recommendations.

```python
m1=df.groupby('Gender')['Quality of Sleep'].mean()
m1.idxmax()
```

```python
plt.figure(figsize=(10,5))
with sns.axes_style('darkgrid'):
  sns.barplot(data=df,x='Gender',y='Quality of Sleep')
  plt.title('Gender VS Quality of Sleep',fontsize=15,color='r')
  plt.show()
```



# **12. Which age group (20-30, 31-40, 41-50, 51+) has the highest prevalence of sleep disorders?**



    * Real-life impact: Supports targeted healthcare interventions.

```python
z1=df['Sleep Disorder']!='None'
df[z1].groupby('Age')['Sleep Disorder'].value_counts(ascending=False)
```

```python

sns.barplot(data=df,x='Age',y='Sleep Disorder',hue='Gender',palette='Set2')
plt.title('Age VS Sleep Disorder',fontsize=15,color='r')
plt.show()
```


# **13. Which occupation gets the least amount of sleep on average?**



    * Real-life impact: Identifies professions vulnerable to sleep deprivation.

```python
m1=df.groupby('Occupation')['Sleep Duration'].mean()
m1.idxmin()
```

```python
m1
```

```python
# Mean sleep duration for each occupation
sleep_avg = df.groupby('Occupation')['Sleep Duration'].mean().reset_index()
sleep_avg
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(
    data=sleep_avg,
    x='Occupation',
    y='Sleep Duration',
    hue='Occupation',
    palette='Set2',
    legend=False
)

plt.title('Average Sleep Duration by Occupation', fontsize=15, color='red')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration (Hours)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# **Conclusion**

The analysis reveals that sleep health is closely connected to lifestyle and physical well-being. Individuals with higher stress levels generally experience lower sleep quality and shorter sleep durations, while those who maintain higher levels of physical activity tend to report better sleep outcomes.

Occupation also plays a significant role in sleep behavior, with some professions showing lower average sleep duration and higher stress levels than others. Additionally, BMI categories and sleep disorders demonstrate important relationships with overall health indicators such as heart rate and blood pressure.

The findings highlight the importance of maintaining a balanced lifestyle that includes regular physical activity, stress management, and healthy sleep habits. These insights can support healthcare professionals, employers, and individuals in making informed decisions to improve overall well-being and reduce potential health risks.

Overall, this project demonstrates how data analysis and visualization can be used to uncover actionable insights from health-related datasets and support data-driven decision-making.
